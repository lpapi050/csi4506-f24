````python
# Obtenir les colonnes qui sont seulement des floats ou integers (sauf la dernière colonne qui est "ID")
columns = data5.select_dtypes(include=['float64', 'int64']).columns[:-1]

# Calculer le nombre de rangées nécessaires pour faire 3 colonnes dans le subplot
num_rows = len(columns) // 3 + (len(columns) % 3 > 0)

# Crée le subplot en spécifiant le nombre de rangées, colonnes et la grosseur de figures + applatir les axes
fig, axs = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, num_rows*4))
axs = axs.flatten()

# Pour chaques colonnes, accèder au subplot à l'index i (axs), crée l'histogramme, ajouter un titre et des x et y labels (j'utilise seaborn pour que les graphs soit plus beau que juste avec matplotlib)
for i, column in enumerate(columns):
    ax = axs[i]
    sns.histplot(data5[column].dropna(), kde=True, ax=ax)
    ax.set_title(f'Distribution de {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Fréquence')

# Delete tout les axs qui sont vide/pas été utiliser, ce qui sont i à la longeur de axs
for j in range(i+1, len(axs)):
    fig.delaxes(axs[j])

# Montrer les histogrammes
plt.tight_layout()
plt.show()
````

5.1 Transformation

Dans mon cas, je veux pas centré les données. Je veux garder la forme de la distribution originale des données. Puisque certaines des valeurs sont très petite comparer à d'autres, ça l'aide acec les valeurs extrèmes. L'autre avantage est que maintenant, toutes les valeurs

````python
# On veut pas run MinMax sur quality
data = data5.drop('quality', axis=1)

# Juste les colonnes numérique
columns = data.select_dtypes(include=['float64', 'int64']).columns

# Déclare notre scaler et ont transforme notre data et crée un dataframe
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data5[columns])
scaled_data = pd.DataFrame(scaled_data, columns=columns)

# Colonne qu'on vas démontrer before and after
column = 'alcohol'

# Crée subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Crée subplot 1
sns.histplot(data5[column], kde=True, ax=axs[0])
axs[0].set_title(f'Distribution de {column} avant transformation')

# Crée subplot 2
sns.histplot(scaled_data[column], kde=True, ax=axs[1])
axs[1].set_title(f'Distribution de {column} après transformation')

# Show subplots
plt.tight_layout()
plt.show()

# Mettre le scaled data back ensemble avec quality
scaled_data = pd.concat([scaled_data, data5['quality']], axis=1)

# Crée nos variables pour X et y qui vas être utiliser plus tard
X_scaled = scaled_data.drop('quality', axis=1)
y_scaled = scaled_data['quality']
````

````python
# OneHot Encodage

# Faire un copie de data5 pour l'exemple
data = data5.copy()

# Declarer l'encodeur et fit quality
encodeur = OneHotEncoder(sparse_output=False)
quality_onehot = encodeur.fit_transform(data[['quality']])

# Pogner les categories et ensuite crée un dataframe avec
categories = encodeur.categories_[0]
noms_colonnes = [f'quality_{int(categorie)}' for categorie in categories]
quality_onehot = pd.DataFrame(quality_onehot, columns=noms_colonnes)

# Concatenate le data originale avec les quality categories OneHot
data = pd.concat([data.drop('quality', axis=1), quality_onehot], axis=1)

print(data.head())
````

6. **Données d'entraînement et cible** : Définissez la variable Python `X` pour désigner les données et `y` pour désigner la classe cible. Assurez-vous de sélectionner uniquement les caractéristiques informatives.


Pour faire certain que nous avons pas de colonnes qui peuvent être exclus ici, nous pouvons checker pour des paires avec hautes corrélation et aussi faire un heatmap qui peut nous aider à visualiser ceci.

J'utilise encore seaborn puisque leurs heatmap est meilleur que celle de base de matplotlib.

Et j'ai aussi réduis les annotations keyword size à 0.6 puisque sinon c'était plus gros que les cases elles même.

````python
y = data5['quality']
X = data5.drop('quality', axis=1)

# Calculer la matrice de corrélation
corr_matrix = data5.drop('quality', axis=1).corr().abs()

# Choisir les paires qui on une haute corrélation (dans ce cas-ci plus haut que 0.9)
high_corr = np.where(corr_matrix > 0.9)
high_corr = [(corr_matrix.index[x], corr_matrix.columns[y]) for x, y in zip(*high_corr) if x != y and x < y]
print("Paires de caractéristiques fortement corrélées :", high_corr)

# Idée de https://github.com/Sudhirr1122/WineQT-Python/blob/main/WineQT.ipynb
sns.heatmap(corr_matrix, annot=True, annot_kws={'size': 6})
````

Si ont trouve des paires avec de hautes corrélation, on enlève une de ces deux colonnes. Dans notre cas, aucune corrélation plus haut que 0.9 sont trouvé. Donc ont a pas besoin d'enlever de colonnes.
